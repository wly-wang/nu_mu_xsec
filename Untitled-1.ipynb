```python
from sklearn.datasets import make_gaussian_quantiles
import matplotlib.pyplot as plt

X, y = make_gaussian_quantiles(
    n_samples=2_000, n_features=10, n_classes=3, random_state=1)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, train_size=0.7, random_state=42
)

from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

weak_learner = DecisionTreeClassifier(max_leaf_nodes=8)
n_estimators = 300
learning_rate = 0.01

adaboost_clf = AdaBoostClassifier(
    estimator=weak_learner,
    n_estimators=n_estimators,
    learning_rate=learning_rate,
    algorithm="SAMME.R",
    random_state=None,
).fit(X_train, y_train)
```

```python
features_test_sig1 = X_test[y_test==1]
features_test_sig2 = X_test[y_test==2]
features_test_bkg = X_test[y_test==0]

y_pred_proba_sig1_train = adaboost_clf.predict_proba(features_test_sig1)
y_pred_proba_sig2_train = adaboost_clf.predict_proba(features_test_sig2)
y_pred_proba_bkg_train = adaboost_clf.predict_proba(features_test_bkg)
bdtScoreSig1Train = y_pred_proba_sig1_train[:,-1]
bdtScoreSig2Train = y_pred_proba_sig2_train[:,-1]
bdtScoreBkgTrain = y_pred_proba_bkg_train[:,-1]
```

```python
plt.hist(bdtScoreSig1Train, 25, color='orange', range=(0.,1.), label=r'Signal - 1', alpha=0.5, histtype='step')
plt.hist(bdtScoreSig2Train, 25, color='darkblue', range=(0.,1.), label=r'Signal - 2', alpha=0.5,histtype='step')
plt.hist(bdtScoreBkgTrain, 25, color='green', range=(0.,1.), label= r'Background - Others', alpha=0.5, density=True, histtype='step')
plt.xlabel("BDT Prediction Score")
plt.ylabel("Area Normalized")
plt.legend(fontsize=8, loc='best', frameon=False)
plt.show()
```
